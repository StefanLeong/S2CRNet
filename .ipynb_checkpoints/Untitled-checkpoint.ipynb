```python
import os
import torch
import argparse
import scripts.datasets as datasets
import scripts.machine as machines
import numpy as np
import random
import seaborn as sns
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
from scripts.models.unet import *
from scripts.models.unet import CreateNetFusionV3
from scripts.utils.imgUtils import im_to_numpy
import imageio
import time
from skimage.metrics import structural_similarity as compare_ssim
from skimage.metrics import peak_signal_noise_ratio as compare_psnr
from skimage.metrics import mean_squared_error as compare_mse
from matplotlib.pyplot import MultipleLocator
import cv2
from thop import profile, clever_format
class objectview(object):
    def __init__(self, *args, **kwargs):
        d = dict(*args, **kwargs)
        self.__dict__ = d

# with open("test_labels.txt") as f:
#     classes = [line.split(',')[0].strip() for line in f.readlines()]
# #     print(classes)

data_config  = objectview({'input_size':2048,
                            'limited_dataset':0,
                            'normalized_input':False,
                            'data_augumentation':False,
                            'base_dir':'/home/ipprlab/datasets/',
                            'preprocess':'scale_width_and_crop',
                            'withLabel':'False',
                            'load_size':286,
                            'crop_size':256,
                            'data':'ABS'})


val_loader = torch.utils.data.DataLoader(datasets.HCOCO('val', data_config), batch_size=1, shuffle=False,
                                             num_workers=0, pin_memory=False, drop_last=True)

res=256
sq = squeezenet1_1(pretrained=True)
model = CreateNetFusionV3(sq, stack=True).cuda()
# model_class = squeezenet1_1(pretrained=True).cuda()
model.load_state_dict(torch.load('/home/ipprlab/projects/xharm/DSCR.tar')['state_dict'])

def CF(img, param, pieces):
    
    # bs x 3 x 1 x 1 x 64
    color_curve_sum = torch.sum(param, 4) + 1e-30
    total_image = img * 0

    for i in range(pieces):
        total_image += torch.clamp(img - 1.0 * i /pieces, 0, 1.0 / pieces) * param[:, :, :, :, i]
    total_image *= pieces/ color_curve_sum
    return total_image


psnrs = []
mses = []
start = torch.cuda.Event(enable_timing=True)
end = torch.cuda.Event(enable_timing=True)


norm = MeanShift([0.485, 0.456, 0.406], [0.229, 0.224, 0.225], norm=True).cuda()

with torch.no_grad():
    for i, batches in enumerate(val_loader):
        if i != 3:
            continue
        inputs = batches['composite_images'].to('cuda')
        target = batches['real_images'].to('cuda')
        mask = batches['mask'].to('cuda')
        fore = batches["fore_images"].to('cuda')  #: self.trans(thumb_foreground),
        foremask = batches["fore_mask"].to('cuda')  # "fore_mask":self.trans(thumb_mask),
        label = batches['label']
        ori_img = batches["ori_img"].to('cuda')
        ori_mask = batches["ori_mask"].to('cuda')
        ori_target = batches["ori_tar"].to('cuda')
        # print(label.shape)
        label_oh = torch.zeros(1, 5).scatter_(1, label.view(label.shape[0], 1), 1).to(
            torch.float32).to('cuda')
        if data_config.withLabel:
            label_oh = label_oh
        else:
            label_oh = None
        print(label_oh)
        feeded = torch.cat([inputs, mask], dim=1).to('cuda')
        fore = torch.cat([fore, foremask], dim=1).to('cuda')
        start.record()
        outputs, param1, param2 = model(ori_img, feeded, fore, label_oh,  data_config.withLabel)
        end.record()
        torch.cuda.synchronize()
        macs, params = profile(model, inputs=(ori_img, feeded, fore, label_oh,  data_config.withLabel))
        macs, params = clever_format([macs, params], "%.3f")

        outputs = outputs if type(outputs) == type(()) else [outputs]

#         f_class = model_class(norm(target * mask))
#         b_class = model_class(inputs * (1-mask))

#         _, index = torch.max(f_class, 1)
    
#         percentage = torch.nn.functional.softmax(f_class, dim=1)[0] * 100

#         print(classes[index[0]], percentage[index[0]].item())
        
#         _, index = torch.max(b_class, 1)
#         percentage = torch.nn.functional.softmax(b_class, dim=1)[0] * 100
#         print(classes[index[0]], percentage[index[0]].item())


        p1 = param1.view(1,3,1,1,64).cpu()

        p2 = param2.view(1,3,1,1,64).cpu()
        xx = np.arange(0,1,1/64)
        print(xx)
#         x_locator = MultipleLocator(24)
#         ax = plt.gca()
#         ax.xaxis.set_major_locator(x_locator)
#         plt.ylim(-7, 7)
#         g=sns.kdeplot(p1[0,0,0,0], shade=True, bw=0.15, color="#098154")
#         plt.xticks([])
#         plt.yticks([])
        plt.ylim(0, 1)
        plt.xlim(0, 1)
        yy = torch.from_numpy(xx).view(1,1,1,64).repeat(1,3,1,1)
        yy = CF(yy,p1,64)

        plt.plot(xx,yy[0,0,0],'r', label='R')
        plt.plot(xx,yy[0,1,0],'g', label='G')
        plt.plot(xx,yy[0,2,0],'b', label='B')
        plt.scatter(xx[10], yy[0,0,0][10], s=25, c='black')
        plt.scatter(xx[25], yy[0,1,0][25],s=25, c='black')
        plt.scatter(xx[45], yy[0,2,0][45],s=25, c='black')
#         plt.text(108, p1[0,0,0,0][108], r'$P_R^{108}$', verticalalignment='bottom', horizontalalignment='right')
#         plt.text(108, p1[0,1,0,0][108], r'$P_G^{108}$', verticalalignment='bottom', horizontalalignment='right')
#         plt.text(108, p1[0,2,0,0][108], r'$P_B^{108}$', verticalalignment='top', horizontalalignment='left')
        plt.vlines(xx[10], 0, yy[0,0,0][10], colors='black', linestyles='dashed', linewidth=1.2)
        plt.vlines(xx[25], 0, yy[0,0,0][25], colors='black', linestyles='dashed', linewidth=1.2)
        plt.vlines(xx[45], 0, yy[0,0,0][45], colors='black', linestyles='dashed', linewidth=1.2)

        plt.hlines(yy[0,0,0][10], 0, xx[10], colors='black', linestyles='dashed', linewidth=1.2)
        plt.hlines(yy[0,1,0][25], 0, xx[25], colors='black', linestyles='dashed', linewidth=1.2)
        plt.hlines(yy[0,2,0][45], 0, xx[45], colors='black', linestyles='dashed', linewidth=1.2)
        plt.ylabel("Output Pixel Value")
        plt.xlabel("Input Pixel Value")
        plt.savefig('/home/ipprlab/projects/xharm/Figs/backbone/' + batches["name"][0].split('.')[0] + '_colorcurve_p1_xy_noline' + '.jpg', dpi=200, bbox_inches='tight')
        
#         plt.figure()
#         plt.xticks([])
#         plt.yticks([])
#         plt.plot(xx,p1[0,0,0,0],'r', linewidth=2.5)
#         plt.savefig('/home/ipprlab/projects/xharm/Figs/backbone/' + batches["name"][0].split('.')[0] + 'colorcurve_r_p1' + '.jpg', dpi=200, bbox_inches='tight')
        
#         plt.figure()
#         plt.xticks([])
#         plt.yticks([])
#         plt.plot(xx,p1[0,1,0,0],'g', linewidth=2.5)
#         plt.savefig('/home/ipprlab/projects/xharm/Figs/backbone/' + batches["name"][0].split('.')[0] + 'colorcurveb_g_p1' + '.jpg', dpi=200, bbox_inches='tight')
        
#         plt.figure()
#         plt.xticks([])
#         plt.yticks([])
#         plt.plot(xx,p1[0,2,0,0],'b', linewidth=2.5)
#         plt.savefig('/home/ipprlab/projects/xharm/Figs/backbone/' + batches["name"][0].split('.')[0] + 'colorcurveb_p1' + '.jpg', dpi=200, bbox_inches='tight')
        plt.figure()
        yy = torch.from_numpy(xx).view(1,1,1,64).repeat(1,3,1,1)
        yy = CF(yy,p2,64)

        plt.plot(xx,yy[0,0,0],'r', label='R')
        plt.plot(xx,yy[0,1,0],'g', label='G')
        plt.plot(xx,yy[0,2,0],'b', label='B')
#         plt.savefig('/home/ipprlab/projects/xharm/Figs/backbone/' + batches["name"][0].split('.')[0] + '_colorcurve_p2_xy_noline' + '.jpg', dpi=200, bbox_inches='tight')
        
       
        print(outputs[0].shape)
        outputs = CF(ori_img, param1.view(ori_img.size(0), ori_img.size(1), 1, 1, 64), 64)
        outputs = CF(outputs, param2.view(ori_img.size(0), ori_img.size(1), 1, 1, 64), 64)

        imfinal = outputs * ori_mask + ori_img * (1-ori_mask)
#         print(imfinal.shape, target.shape)
        img_show = imfinal.squeeze(0).cpu().permute(1, 2, 0).numpy()

        imageio.imwrite( 'Input' + '.jpg', ori_target.squeeze(0).cpu().permute(1, 2, 0).numpy())

#         imageio.imwrite( 'result' + '.jpg', img_show)

#         psnr = compare_psnr(imfinal, target, data_range=imfinal.max() - imfinal.min())
#         ssim = compare_ssim(imfinal, target, multichannel=True)
    
# #             imfinal = outputs * ori_mask + ori_img * (1-ori_mask)
# #             img_show = imfinal.squeeze(0).cpu().permute(1, 2, 0).numpy()
        imfinal = im_to_numpy(torch.clamp(imfinal[0] * 255, min=0.0, max=255.0)).astype(np.uint8)
        ori_target = im_to_numpy(torch.clamp(ori_target[0] * 255, min=0.0, max=255.0)).astype(np.uint8)

        psnr = compare_psnr(imfinal, ori_target, data_range=imfinal.max() - imfinal.min())
        mse = compare_mse(imfinal, ori_target)
        
        print(psnr, mse )
        print(macs, params, start.elapsed_time(end))


#         plt.savefig(str(i)+".jpg")
       
            
```
