```python
import torch
import torch.nn as nn
from tensorboard.backend.event_processing import event_accumulator as ea
import csv
import matplotlib.pyplot as plt
import numpy as np
def readcsv(files):
    csvfile = open(files, 'r')
    plots = csv.reader(csvfile, delimiter=',')
    x = []
    y = []
    for row in plots:
#         print(row[1])
        x.append(int(row[1]))
        y.append(float(row[2]))
    return x, y

plt.figure()

x1, y1 = readcsv('s2am.csv')
plt.plot(x1, y1, label='S$^2$AM', linewidth=2)

x2, y2 = readcsv('rainnet.csv')
plt.plot(x2, y2, label='RainNet', linewidth=2)
x3, y3 = readcsv('ours.csv')
plt.plot(x3, y3, label='Ours', linewidth=2)
# plt.ylim([0, 1])
plt.xlim([0, 50])
plt.tick_params(labelsize=12)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('PSNR(dB)',fontsize=12)
plt.legend()
# plt.grid()
y_ticks_1 = np.arange(25, 38, 2)
        #             ax1.set_ylabel('Style Parameter Values')
plt.yticks(y_ticks_1)
plt.legend(fontsize=12)
# plt.show()
plt.savefig('/home/ipprlab/projects/xharm/Figs//' + 'converge_speed' + '.pdf', dpi=200, bbox_inches='tight')

```
