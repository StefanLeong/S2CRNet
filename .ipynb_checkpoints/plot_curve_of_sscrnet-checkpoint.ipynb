```python
import os
import torch
import argparse
import sys


sys.path.append('..')
sys.path.insert(0,'..')

import scripts.datasets as datasets
import scripts.machine as machines
import numpy as np
import random
%matplotlib inline
import matplotlib.pyplot as plt
from scripts.models.unet import *
from scripts.models.backbone_unet import *
from scripts.utils.imgUtils import im_to_numpy
import imageio
import time
from skimage.metrics import structural_similarity as compare_ssim
from skimage.metrics import peak_signal_noise_ratio as compare_psnr
from skimage.metrics import mean_squared_error as compare_mse

class objectview(object):
    def __init__(self, *args, **kwargs):
        d = dict(*args, **kwargs)
        self.__dict__ = d

data_config  = objectview({'input_size':256,
                            'limited_dataset':0,
                            'normalized_input':False,
                            'data_augumentation':False,
                            'base_dir':'/data/Datasets/iHarmony',
                            'data':'HFlickr'})


val_loader = torch.utils.data.DataLoader(datasets.HCOCO('val', data_config), batch_size=1, shuffle=True,
                                             num_workers=0, pin_memory=False, drop_last=True)

def CF(img, param, pieces):
    
    # bs x 3 x 1 x 1 x 64
    color_curve_sum = torch.sum(param, 4) + 1e-30
    total_image = img * 0

    for i in range(pieces):
        total_image += torch.clamp(img - 1.0 * i /pieces, 0, 1.0 / pieces) * param[:, :, :, :, i]
    total_image *= pieces/ color_curve_sum
    return total_image

model = fusionv3s().to('cuda:0')
model.load_state_dict(torch.load('/data/ImageHarmonization/splitNet/eval/ss_sspp_HFlickr_fusionv3s/model_best.pth.tar')['state_dict'])

res=(256,256)

plt.figure(figsize=(25,5))
psnrs = []
mses = []
with torch.no_grad():
    for i, batches in enumerate(val_loader):

        inputs = batches['composite_images'].to('cuda:0')
        target = batches['real_images'].to('cuda:0')
        mask = batches['mask'].to('cuda:0')      
        fore = batches["fore_images"].to('cuda:0') #: self.trans(thumb_foreground),
        foremask = batches["fore_mask"].to('cuda:0')  # "fore_mask":self.trans(thumb_mask),
       
        feeded = torch.cat([inputs, mask], dim=1).to('cuda:0')
        fore = torch.cat([fore, foremask], dim=1).to('cuda:0')
        s2,s1,p2,p1 = model(feeded, fore)
        
        imfinal2 = s2 * mask + inputs * (1-mask)
        imfinal1 = s1 * mask + inputs * (1-mask)

        
        imall = torch.cat([inputs,mask.repeat(1,3,1,1),imfinal1,imfinal2,target],dim=3)
        imfinal = im_to_numpy(torch.clamp(imall[0] * 255, min=0.0, max=255.0)).astype(np.uint8)
        
        plt.imshow(imfinal)
        
        
        plt.figure()
        p1 = p1.view(1,3,1,1,64)
        p2 = p1.view(1,3,1,1,64)
        xx = np.arange(0,1,1/64)
        yy = torch.from_numpy(xx).view(1,1,1,64).repeat(1,3,1,1).to('cuda:0')
        
        yy = CF(yy,p1,64)
        plt.plot(xx,yy[0,0,0].cpu(),'r')
        plt.plot(xx,yy[0,1,0].cpu(),'g')
        plt.plot(xx,yy[0,2,0].cpu(),'b')
        
        plt.figure()
        yy = CF(yy,p2,64)
        plt.plot(xx,yy[0,0,0].cpu(),'r')
        plt.plot(xx,yy[0,1,0].cpu(),'g')
        plt.plot(xx,yy[0,2,0].cpu(),'b')
        
        
#         plt.figure()
#         p = p2.view(1,3,1,1,64).cpu()
#         xx = np.arange(0,1,1/64)
#         plt.plot(xx,p[0,0,0,0],'r')
#         plt.plot(xx,p[0,1,0,0],'g')
#         plt.plot(xx,p[0,2,0,0],'b')
        break;
        



#         psnr = compare_psnr(imfinal, target, data_range=imfinal.max() - imfinal.min())
#         mse = compare_mse(imfinal, target)
        
#         psnrs.append(psnr)
#         mses.append(mse)

# print(np.mean(psnrs), np.mean(mses))

#         imall = torch.cat([inputs,mask.repeat(1,3,1,1),imfinal,target],dim=3)
#         imfinal = im_to_numpy(torch.clamp(imall[0] * 255, min=0.0, max=255.0)).astype(np.uint8)
        
#         imageio.imwrite('Hday2night/'+batches['name'][0], imfinal)
            
```
