```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)

x = [0.62, 18.98, 47.2, 21.71, 18.08]
y = [36.04, 34.75, 34.35, 35.86, 37.08]
s = [0.72, 54, 66.7, 11.59, 38.5]
n = ['Ours\n(0.72M)', 'DoveNet\n(54M)', 'S2AM\n(66.7M)', 'BMVC21\n(11.59M)', 'WACV21\n(38.5M)']
colors =  np.random.rand(len(n))
s = [ 10*i for i in s]

plt.scatter(x,y,s=s,c=colors, alpha=0.5)
plt.scatter(x,y,s=[1,1,1,1,1],c=[0.4,0.4,0.4,0.4,0.4])


for i, txt in enumerate(n):
    plt.annotate(txt, (x[i], y[i]))

plt.xscale('log')
plt.xlim(0,200)
plt.ylim(34.0,38.0)
plt.xlabel('GMacs')
plt.ylabel('PSNR')
plt.grid(True)
plt.savefig('gmac.png',dpi=300,bbox_inches='tight')
```
